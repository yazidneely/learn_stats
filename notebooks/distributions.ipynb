# 1.0 - Discrete Probabilities

## 1.1 Uniform Distribution

The uniform distrubtion describes the probability mass function a discrete random variable where all outcomes are equal probable.<br>

$$P(X)= \frac{1}{n}, \forall x : [a,b]$$
$$\mu = \frac{a+b}{2}$$
$$\sigma^2 = \frac{(b-a+1)^2-1}{12}$$

For instance rolling a fair six sided die would result in the following:
$$P(X) = \frac{1}{6}, \forall x : [1,6]$$

The pmf is of this random variable shown in the following graph:

```python
from scipy.stats import randint

# Define the range of the discrete uniform distribution (e.g., a six-sided die)
low = 1
high = 6

# Create a randint object for the discrete uniform distribution
# Note: high in randint is exclusive, so we add 1 to include the upper bound
discrete_uniform_dist = randint(low, high + 1)

# Generate an array of possible values (x-axis)
values = np.arange(low, high + 1)

# Calculate the PMF for each value
pmf_values = discrete_uniform_dist.pmf(values)

# Plot the PMF
plt.figure(figsize=(8, 5))
plt.bar(values, pmf_values, width=0.5, edgecolor='black', alpha=0.7)
plt.title('PMF of a Discrete Uniform Distribution')
plt.xlabel('Possible Values')
plt.ylabel('Probability')
plt.xticks(values) # Ensure integer ticks on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

result=pd.Series(randint.rvs(low,high+1,size=10000))
result.describe()
```

## 1.2 - Poisson Distribution

Probability Mass Function:
$$ P(X=k)=\frac{\lambda^ke^{-\lambda}}{k!}, \lambda \in [0,\infty)$$
$$ \mu = \lambda$$
$$ \sigma^2 = \lambda$$

```python
from scipy.stats import poisson
k=np.linspace(0,20,21)
thelambda = 7
pmf_values = poisson.pmf(k, thelambda)

plt.figure(figsize=(8, 5))
plt.bar(k, pmf_values, width=0.5, edgecolor='black', alpha=0.7)
plt.title('PMF of a Poisson Distribution: Lambda = '+str(thelambda))
plt.xlabel('Possible Values')
plt.ylabel('Probability')
plt.xticks(k) # Ensure integer ticks on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
result=pd.Series(poisson.rvs(thelambda,size=10000))
result.describe()
```

# 2.0 - Continous Distributions

## 2.1 - Normal Distribution

$$ P(X<=x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

```python
from scipy.stats import norm
mu = 3
std = 2
x=np.linspace(-3,9,1000)

pdf_values = norm.pdf(x,loc=mu,scale=std)

plt.plot(x,pdf_values)
plt.title('PDF of a Normal Distribution: mean = 3, std = 2')
plt.xlabel('Possible Values')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

result=pd.Series(norm.rvs(loc=mu,scale=std,size=10000))
result.describe()
```
